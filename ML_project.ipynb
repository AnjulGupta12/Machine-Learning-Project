```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species") \
   .map(plt.scatter, "petal_length", "petal_width")
plt.show()
   # FacetGrid : https://seaborn.pydata.org/generated/seaborn.FacetGrid.html
   # facegrid is used to give different different colour 
   # whitegrid for white background
   # facetgrid for different colour 
   # hue : according to 'species'
```

Change the set style in the above plot. Also how will you figure out which color is for which specie.

```python
sns.set_style("darkgrid")
sns.FacetGrid(iris, hue="species") \
   .map(plt.scatter, "petal_length", "petal_width").add_legend()
plt.show()
```

**Pair Plots**

```python
# Pair Plot
sns.set_style("whitegrid")
sns.pairplot(iris, hue="species", height=2, diag_kind="hist").add_legend()
plt.show()
#https://seaborn.pydata.org/generated/seaborn.pairplot.html

#this same x-y axis graph(diag_kind="hist") is a histogram, that tells frequency of data... this way this one is also useful
```
